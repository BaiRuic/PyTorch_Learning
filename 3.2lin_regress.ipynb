```python
import numpy as np
from matplotlib import pyplot as plt
import torch
import random
from IPython  import display
%matplotlib inline
```

```python
## 构造数据集
num_examples = 1000
num_input = 2
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features = torch.randn(num_examples, num_input, dtype = torch.float32)
labels = features.mv(true_w.t())  # 矩阵 乘 向量
labels += torch.tensor(np.random.normal(0,0.01,size=labels.shape), dtype = torch.float32)
```

```python
# 读取数据集
def data_iter(batch_size,features,labels):
    num_examples = len(features)
    indices  = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0,num_examples, batch_size):
        index = torch.LongTensor(indices[i: min(i+batch_size,num_examples)])  # 将python类型数字转化为 tensor类型
        yield  features.index_select(0, index), labels.index_select(0, index)       # 0 表示按照行索引 j表示索引那些行or列  为torch.tensor 向量类型  
```

```python
## 初始化模型参数
w = torch.tensor(np.random.normal(0,0.01,(num_input,1)), dtype = torch.float32, requires_grad=True)
b = torch.zeros(1,dtype=torch.float32, requires_grad=True)
```

```python
## 定义模型
def linreg(X, w, b):
    return torch.mm(X, w) + b
```

```python
## 定义损失函数
def squared_loss(y_hat, y):
    return (y_hat - y.view(y_hat.shape)) **2 / 2
```

```python
# 定义优化函数
def sgd(params, lr, batch_size):
    for param in params:
        param.data -= lr * param.grad / batch_size
```

```python
# 训练模型
batch_size = 10
lr = 0.01
num_epochs = 5
net = linreg
loss = squared_loss

for epoch in range(num_epochs):
    for X,y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y).sum()
        l.backward()
        sgd([w, b], lr, batch_size)
        
        w.grad.data.zero_()
        b.grad.data.zero_()
    train_l = loss(net(features, w, b), labels)
    print('epochs %d, loss %f' %(epoch+1, train_l.mean().item()))
    
```
